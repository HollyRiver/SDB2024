### 데이터베이스 설계

전체 흐름

`-` 1단계 : 요구 사항 분석.

> 데이터베이스가 요구하는 목록을 분석

사용자의 요구 사항을 정확히 분석해야 재수정, 재개발의 비용이 절감됨.

ex)

* 회원가입에는 아이디, 비밀번호, 이름, 나이, 직업을 입력해야 함. > 회원에 대한 개체 정보(속성) & 기본키의 대략적인 후보
* 가입한 회원에게는 등급과 적립금이 부여(관계를 굳이 만들어야 할 필요는 없어보임) / 회원은 회원 아이디로 식별(<span style="background-color:red">기본키</span>)
* 상품에 대한 상품번호, 상품명, 재고량, 단가정보를 유지(상품이라는 개체와 그의 속성) / 상품은 상품번호로 식별(<span style="background-color:red">기본키</span>)
* 회원은 여러 상품의 주문이 가능하며 하나의 상품을 여러 회원이 주문할 수 있음(다대다 관계)
* 회원이 상품을 주문하면 주문번호, 주문수량, 배송지, 주문일자 정보를 유지(관계의 속성)
* 각 상품은 한 제조업체가 공급하고, 제조업체 하나는 여러 상품을 공급할 수 있음(제조업체라는 개체와, 제조업체-상품의 일대다 관계)
* 제조업체가 상품을 공급하면 공급일자와 공급량 정보를 유지(관계의 속성) / 제조업체명, 전화번호, 위치, 담당자 정보 유지(제조업체의 속성)
* 제조업체는 제조업체명으로 식별(<span style="background-color:red">기본키</span>) / 회원은 게시글을 여러 개 작성, 게시글 하나는 한 명의 회원만 작성(회원-게시글 간 일대다 관계)
* 게시글에 대한 글번호, 글제목, 글내용, 작성일자 정보 유지(게시글이라는 개체와 그 속성) / 게시글은 글번호로 식별(<span style="background-color:red">기본키</span>)

> 개체와 그 속성, 개체 관 관계성을 파악

**개체와 속성 추출**

> 개체는 물리적 실체, 개념적 실체가 될 수 있음. 명사 중심으로 추출.

|||
|:-:|:-:|
|회원|회원아이디, 비밀번호, 이름, 나이, 직업, 등급, 적립금|
|상품|상품번호,상품명, 재고량, 단가|
|제조업체|제조업체명, 전화번호, 위치, 담당자|
|게시글|글번호, 글제목, 글내용, 작성일자|

**관계 추출**

> 관계를 맺더라도, 그것에서 추가적인 정보가 존재하지 않을 수 있음. 속성이 없는 관계가 있을 수 있다.(단순히 매핑으로서만 존재) 동사 중심으로 추출.

|관계|참여 개체|관계성|속성|
|:-:|:-:|:-:|:-:|
|주문|회원 (선택) / 상품 (선택) > 회원은 안살수도 있고, 상품은 안팔릴수도 있음.|다대다|주문번호, 주문수량, 배송지, 주문일자|
|공급|상품 (필수) / 제조업체 (선택) > 공급이 되어야만 상품이 존재함.|일대다|공급일자, 공급량|
|작성|회원 (선택) / 게시글 (필수) > 회원은 게시글을 쓸 수도 있고, 작성해야만 게시글은 존재할 수 있음.|일대다|-|

`-` 3단계 : 논리적 설계

> 릴레이션 설계

개체는 모두 릴레이션을 만들고, 관계는 만들 수도 있다. 만들지 않고 개체의 릴레이션에 끼워넣을 수도 있다.

* 관계에는 포함하고 있는 정보 뿐만 아니라, 관계를 맺는 각 개체가 가지고 있는 속성도 가져온다.(외래키?) 즉, 관계의 속성(없을 수도 있음) + 개체별 기본키 2개

> 다대다의 경우 튜플이 구분이 되기 위해선 릴레이션을 별도로 만들어야만 할듯. (주문)

> 일대다의 경우 관계를 개체에 묶어서 저장하고, 외래키를 사용하여 관계를 표현한다. 일대다니까, 다의 방향에 매핑에 대한 정보를 저장해야만 모든 매핑 정보를 유지할 수 있을듯. (게시글 -> 작성 / 상품 -> 공급. 외래키로 참조.)

* 기본 규칙 5가지를 이용하며, 다음 정규화 과정을 거쳐 완성하게 된다.

1) 모든 개체의 릴레이션 변환
2) 다대다 관계의 릴레이션 변환
3) 일대다 관계의 외래키 표현
4) 일대일 관계의 외래키 표현(여긴 아무데나 붙여도 될듯)
5) 다중 값 속성의 릴레이션 변환(?)

**1. 모든 개체는 릴레이션으로 변환**

**2. 다대다 관계의 릴레이션 변환**

> 고객과 상품(다대다) : 두 개의 아이디를 튜플로 엮어 기본키로 삼거나, **새로 인덱스를 추가하여 기본키로 사용한다.** $\to$ 주문번호
>
> 만약에, 주문일자라는 열을 추가한다면, 고객번호와 상품번호를 묶은 튜플을 기본키로 설정하는 것에 문제가 있지 않을까???

**3. 일대다 관계의 외래키 표현**

속성을 가지는 일반 개체의 경우

> 다에 해당하는 곳에 속성을 추가(관계의 속성과 참조하는 외래키)

속성을 가지지 않은 약한 개체의 경우

> 똑같음.

**4. 일대일 관계의 외래키 표현**

> 어느 쪽에 붙이든 튜플의 개수는 바뀌지 않는다. 그러므로 조금 더 정보가 많은 쪽으로 붙여준다. 외래키와 없는 자료의 표기를 NULL로 잘 해줘야 한다.
>
> 또는 둘 다 필수 참여 개체인 경우에 양 측 정보를 모두 포함하도록 릴레이션을 변환한다.

**5. 다중 값 속성의 릴레이션 변환**



**특별한 목적이 있지 않는 한, 일대일과 일대다 관계에서는 릴레이션을 만들지 않는다.**

### 정규화

어떤 한 줄을 지웠는데, 수정했는데, 연결되는 애들을 전부 쫓아가서 수정해야 하는 상황은 일어나지 않았으면 좋겠다.

> 단순 작업의 반복을 하는 것을 원하지 않는다.



> 고객이름이 "정소화"인 튜플을 제거하고자 할 때, 세 개의 튜플을 없애야 한다. 이 경우 고객이름과 등급을 따로 떼어서 릴레이션으로 만들고, 이후 조인으로 처리.
>
> 릴레이션이 너무 많으면 또 조인을 많이 해야하기 때문에 복잡해진다는 단점이 있음.

이상현상 : 고객이름이나 등급으로 관리를 하면 발생할 수 있음. 하지만 이벤트번호에 대해서는 이상현상이 발생하지 않는다. 모든 속성에 대해서 이상현상이 발생하지 않게 하려면 더욱 복잡한 정규형 과정을 거쳐야 함.

정규화는 논리적 설계 이후에 진행한다 : 발표할 때에 정규화로 조인하는 법도 해야 한다.
