### 1.  데이터베이스의 정의

※ 여러 사용자가 **공유**하여 사용할 수 있도록 **통합**하여 **저장**한 **운영데이터**의 집합

* 공유데이터 : 여러 사용자가 동시에 사용이 가능
* 통합데이터 : 데이터의 중복은 허용하지 않거나 최소화
* 저장데이터 : 컴퓨터로 접근 가능한 저장방식 사용
* 운영데이터 : 조작 및 이용을 위한 지속성 존재

### 2. 데이터베이스의 기본 특징

* 실시간 접근성 : 언제나 접근이 가능함
* 계속 변화 : 내용이 계속 수정, 업데이트
* 내용 기반 참조 : 내용을 기반으로 데이터 탐색이 가능함
* 동시 공유 : 여러 사용자가 동시에 사용이 가능

### 3. DBMS를 사용하는 이유

기존 파일 시스템의 단점 해소를 위해

* 내용의 중복(튜플의 유일성) : 저장공간이 낭비되고, 일관성 유지가 어려울 수 있음.
* 무결성 유지의 어려움 : 개체 무결성 제약 / 참조 무결성 제약. 

> 데이터가 자주 바뀌지 않거나, 소규모인 경우 DBMS가 없는 것이 나을 수도 있다.

`-` 데이터 언어

* 데이터 정의어 : 스키마 정의 및 수정, 테이블 객체 정의
* 데이터 조작어 : 데이터 접근, 삽입, 삭제, 수정, 검색
* 데이터 제어어 : 응용 프로그램 접근 권한 설정, 내부 규칙 정의

### 4. 제 2세대 DBMS

데이터베이스 구조 : 외부 단계 / 개념 단계 / 내부 단계 $\star\star$

* 개념 단계 : 전체 관리자 관점에서의 단계
* 외부 단계 : 사용자에게 노출될 단계
* 내부 단계 : 실제 저장장치 관점에서의 단계

> 개념 단계와 외부 단계에서 사용하는 이름(스키마)이 서로 달라도 된다. 단, 상호 공유되는 데이터는 매핑이 되어야 한다.

`-` 데이터 독립성 : 개념 단계의 스키마는 고정하되, 외부 스키마는 유동적으로 변화시킬 수 있다. 이 때, 매핑으로 각 단계의 독립성을 유지한다.

> 사상에 대한 정보 : 데이터 사전(Data dictionary) / 메타 데이터(Meta Data)

### 5. 데이터베이스 사용자

* 데이터베이스 관리자 : DBMS를 운영 및 관리(개념 / 내부 단계 관리 운영 및 보안 관리)
* 최종 사용자(일반 사용자) : 데이터를 이용하거나 분석에 활용
* 응용 프로그래머 : DB와 연동되는 응용 프로그램 작성, 알고리즘 제작

### 6. 데이터 언어(내부 단계)

* 데이터 정의어(스키마 정의 및 삭제 / 수정 -> 데이터 사전 내 저장) / 데이터 조작어 / 데이터 제어어(무결성과 일관성 유지에 필요한 내부 규칙 정의)

### 7. 개체-관계 모델

> 개체(entity)와 속성(attribute), 인스턴스(instance)로 정의되고, 개체 간 관계를 중시

* 속성의 분류
    > * 단일 값 속성 / 다중 값 속성 : 여러 의미의 정보를 가질 수 있느냐의 여부 (이름 / 연락처 : 연락처의 경우 집전화 / 개인전화 등 여러 정보가 혼용이 가능)
    > * 단순 속성 / 복합 속성 : 쪼갤 수 있는지 여부 (이름 / 생년월일 : 년, 월, 일로 쪼갤 수 있음)
    > * 유도 속성 : 기존 속성 값에서 다른 속성의 값이 반드시 유도되는 것 (책이라는 엔티티에서의 판매가격, 가격과 할인률이라는 두 개의 속성에 의하여 유도된다.)
    > * 널 속성
    > * 키 속성
    
* E-R 다이어그램

|값|표기|
|-|-|
|개체||
|관계||
|단일 값 속성 + 단순 속성||
|다중 값 속성||
|복합 속성(생년월일)||
|유도 속성(판매 가격)||
|키 속성||
|관계의 참여 특성 구분| 고객 전체가 구매하였으므로 <전체참여>, 책은 일부만 팔렸으므로 <선택적 참여>|
|관계의 종속성 구분| 부양이란 관계는 부양가족이 필수적이므로 <완전종속>|
|개체 간 관계의 개수| 고객은 여러 책을 구입할 수 있고, 책도 여러 고객에게 팔릴 수 있음. 다대다 관계 / 한 출판사에서 출간한 책은 다른 출판사에서는 출간할 수 없음. 일대다 관계|

### 8. 논리적 데이터 모델

> 개체-관계 모델의 개념적 구조를 어떻게 데이터베이스에 저장할 것인지 논리적으로(logical) 표현한 것

### 9. 관계 데이터 모델

개체와 관계 데이터 테이블을 따로 저장

> 한 개체에 대한 데이터를 릴레이션 하나에 담아 데이터베이스에 저장하는 것. 릴레이션 / 튜플-속성

* 도메인 : 한 속성이 가질 수 있는 모든 값의 집합(벡터공간)
* 널 / 카디널리티 및 차수 : 정해지지 않은 값 / 튜플의 개수(카디널리티 n), 속성의 차원(차수 p)

`-` 릴레이션의 특징 $\star\star$

1. 튜플의 유일성 : 각 튜플은 다른 튜플들과 중복되지 아니한다. $\to$ 키를 겹치지 않게 하고, 겹치는 것이 있으면 없앰
2. 튜플의 무순서 / 속성의 무순서 : 행이나 열의 순서를 교환(commutation)해도 동일한 릴레이션임
3. 속성의 원자성 : 속성은 논리적으로 분해가 불가능함

`-` 키의 종류

* 슈퍼키 : 튜플의 유일성을 구분지을 수 있는 키
* 후보키 : 튜플의 유일성 및 차원의 최소성 만족
* 기본키 : 후보키 중 기본으로 사용되는 하나의 키
* 대체키 : 기본키로 선택되지 못한 후보키

> 슈퍼키 ⊃ 후보키 ∋ 기본키 / 후보키 － 기본키 = 대체키

* 외래키 : 릴레이션들 간의 관계에서 파생되는 키, 기본키와 도메인이 같으며, 중복이 허용되고, 널값을 가질수도 있다(참조 무결성 제약).



> "user id"를 외래키로 받아 누가 주문하는 지 알 수 있도록 하였다.

`-` 데이터 모델 표시 요소 : 데이터 구조(E-R모델, 스키마), 제약조건(일관성, 무결성), 연산(쿼리)

`-` 관계 데이터 모델의 제약

* 개체 무결성 제약 : 기본키를 구성하는 모든 속성은 널값을 가지면 안된다.
* 참조 무결성 제약 : 외래키는 참고하는 기본키와 도메인이 같거나, NULL이여야 한다.

`-` 무결성

* 개체 무결성 / 참조 무결성 / 도메인 무결성 준수
* 데이터 구조와 응용프로그램의 독립성

### 10. 비정형 데이터베이스

Non-SQL 데이터베이스

> 관계 데이터베이스가 아닌 객체형 데이터베이스

* 통합 / 일관성이 깨지는 경우가 많음. 용량도 많고 규격이 복잡해 관계 데이터베이스로 구성이 불가능하다.
* 취약성과 데이터 보존을 위해 여러 데이터베이스에 분산하여 저장한다. 이 경우 싱크가 맞지 않아 일관성이 깨질 수 있다.

`-` CAP

> Consistency : 일관성 / 일치성, 모든 사용자는 같은 데이터를 이용
>
> Availiability : 접근성, 모든 사용자의 R/W 가 가능하고 장애 노드를 제외한 모든 노드에 접근이 가능
>
> Partition Tolerance : 물리적 네트워크가 분할되어도 각 노드는 시스템 작동이 가능해야 함

* 핵심정보는 CA를 만족하는 관계형 DBMS에서, 막대한 양의 데이터는 P를 만족하는 비정형 DBMS에서 취급

`-` 종류

* Column / Graph : 무식하게 2차원 행렬로 펴서 저장하여 널값이 생길 수 있음, 하나의 행에 길이가 다른 여러 개의 열을 가질 수 있음 / 각각의 개체와 관계를 저장, 관계 노드를 직접 연결해 관리가 어려움
* Key-value / Document : Json, Xml / 트리형 자료구조
* Mongo DB에서는 Relation = Collection / Tuples = Documents / Attributes = Fields라고 함
> Replica set : P-S-S(Secondary set이 두개) / P-S-A(Arbiter set이 존재)

### 11. 관계 데이터 연산

`-` 관계 대수

* 일반 집합 연산자 : 합, 교, 차, 카티션 프로덕트
* 순수 관계 연산자 : Select($\sigma_{Restrictions}(R)$), Project($\pi_{Attributes}(R)$), Join($R \Join S$), Division($R ÷ R$)
> 카티션 프로덕트를 제외한 일반 집합 연산에서는 각 릴레이션의 속성과 스키마가 같아야 함.
>
> 순수 관계 연산자 중 Select에서 조건은 ∧ ∨ ¬ (and or not)으로 표기하며, Division의 경우 한쪽만 가져오되 인덱스 열은 제거한다.
  * Semi-Join : 조인을 하되 뚫려있는 삼각형의 정보는 날려버린다.
  * Outer-Join : 조인을 하되 뻗어있는 방향의 정보는 하나도 버리지 않는다.
