### 비정형 데이터베이스

Non-SQL 데이터베이스.

> 관계 데이터베이스가 아닌 객체형 데이터베이스
>
> 비정형데이터는 복잡한 형태를 가지고 있고, 용량이 많아 관계 데이터베이스로 구성이 불가.
>
> 비중복성 / 일관성을 지키지 못함. 그냥 용량만 확보해서 데이터를 쌓아두자...

취약성이 커짐. 데이터가 소멸될 수도 있음. -> 데이터의 중복을 허용, 가장 가까운 데이터를 취득

> 수정할 때 여러 데이터베이스를 수정해야 함, 따라서 싱크가 맞지 않아 일관성이 깨질 수 있음.
>
> 파일의 용량이 크기 때문에 통신이 오래 걸림.

* 분산 시스템의 등장 : 데이터 저장을 한 곳이 아니라 여러 곳에 나누어 하는 방식

* CAP를 동시에 만족시키는 것은 어려움

> Consistency : 모든 노드 데이터 일치성, 모든 사용자는 같은 데이터를 이용
>
> Availability : 모든 사용자의 R/W가 가능. 장애 노드를 제외한 모든 노드에 접근
>
> Partition Tolerance : 물리적 네트워크 분할에도 각 노드는 시스템 작동이 가능해야 함.

핵심 정보는 CA를 만족하는 관계형 DBMS에서 / 막대한 양의 데이터에 대해서는 P를 만족하는 비정형 DBMS에서


`-` 비정형 데이터 : 이미지, 텍스트, 음성, 네트워크

> 수량화 할 수 있느냐? 소스 원천의 수준에서는 수량화가 안되는 것이 있으나(디지털의 한계), 생성이 된 시점에서는 수량화할 수 있다.

image : RGB 값으로 저장
> 3차원 텐서 구조로 저장된다.

text : one-hot 방식의 벡터로 저장
> 단어가 어떠한 형태소 위치에 존재하면 1, 나머지는 0으로 저장. 6만~8만 단어가 있다고 하면 길이가 그만큼인 벡터가 됨...
>
> 길이가 가변적이고, 중간중간 넣어줘야 할 정보가 많음. 텐서 구조로 저장된다.

sound : 파형으로 표현이 가능
> 시간축에 따른 음성의 세기로 표현, 텐서 구조

network : 연결되었다면 1, 아니면 0인 행렬구조

**일렬 벡터가 아닌 텐서로 저장되는 경우가 많음. 차원 구조를 깨는 것이 데이터를 유실시킬 수 있음. 따라서 테이블 형태로 저장되는 것이 어렵다.**
