### 개체-관계 모델

네모가 개체, 동그라미가 속성, 개체의 개별 값이 인스턴스

> 고객, 고객아이디, hcssk2800. 개체에 여러 속성(열)이 있고, 그 실제 값들이 인스턴스(행)

* 널 속성 null attribute : 결정이 안되거나 모르는 값
* 키 속성 key attribute : 개체 집합에 존재하는 각 개체 인스턴스를 식별하는 데 사용

`-` 관계 : 개체와 개체 간의 연관성

> 관계가 모두 맺어지지 않고 부분적으로 맺어져도 된다.

* 이항(개체 타입 두 개), 삼항(개체 타입 세 개), 순환관계(인스턴스의 값이 정해진 것으로 바뀌는 것, 재귀)
* 개체가 가질 수 있는 값 전체에 대한 매핑 : 전체 참여(두 줄 표시), 부분만 매핑 : 선택적 참여(한 줄 표시)

`-` 관계의 종속성 existence dependence : 하나의 개체가 없어질 경우 같이 없어지는 종속 관계

* 부모와 자녀 : 강한 개체-약한 개체 간의 일대다 관계 / 직장인과 부양가족 : 강한 개체-약한 개체 간의 일대 다 관계
* key attribute : 이름은 동명이인이 존재하므로 (직원, 부양가족)으로 묶어 인스턴스를 구분하는 것이 바람직함. > 부양가족 이름 : 구분자 / 부분키(partial key)



개체는 3개, 관계는 2개, 고객 아이디는 key attribute, 구매라는 관계의 속성은 2개, 고객과 책의 관계는 다대다, 책과 출판사의 관계는 일대다


가 : a,b,c,d -> 김철수 : a×3. 다대다에서 몇 권 이상을 살 수도 있음(수량을 따로 명시해야 할 필요가 있음.)

### 논리적 데이터 모델

개체-관계 모델의 개념적 구조를 어떻게 데이터베이스에 저장할 것인지 논리적으로 표현한 것.

* 계층 데이터 모델 : 트리 구조 기반
> 데이터베이스의 논리구조가 그래프 내지 네트워크.
>
> 일대다만 표현이 되어 다대다 관계 표현을 위해 관계를 추가해야.
>
> 구조가 복잡해지면 데이터 삽입-삭제-수정 연산과 데이터 검색이 어려워짐

직관적이나, 운영이 어려움



> 개체와 관계들을 따로 테이블로 구분하여 관리하는 게 좋음.

### 관계 데이터 모델

개체와 관계 데이터 테이블을 따로 저장.

**한 개체에 대한 데이터를 릴레이션(테이블) 하나에 담아 데이터베이스에 저장하는 것.**

> key attribute가 개체간 관계를 구분해줌 | 릴레이션

릴레이션 relation. 개체도 릴레이션, 관계도 릴레이션...

* 속성 attribute / 튜플 tuple(릴레이션의 인스턴스)
* 도메인 : 속성이 가질 수 있는 모든 값의 집합(벡터공간)
* 널 / 차수 및 카디널리티 : 정해지지 않은 값 / 속성의 차원(널 p), 튜플의 개수(카디널리티 n)



> key attribute 관리
>
> 고객아이디를 무조건 다르게 설정하고 키로 지정 + 고객이름을 partial key로 묶어 지정

`-` 릴레이션의 특징

1. 튜플의 유일성 : 튜플 하나하나는 다른 원소와 겹치지 않아야. -> 기존과 같은 게 있으면 원천적으로 입력되지 않게 만듦 / 주기적으로 계속 확인 / 키를 겹치지 않게 하고 겹치는 것이 있으면 없앰. key attribute 관리가 중요. 겹치게 되면 용량이 감당되지 않는다.
2. 튜플의 무순서 / 속성의 무순서 : 행이나 열의 순서를 바꿔도 같은 릴레이션임 commutation(교환)해도 동일함.
3. 속성의 원자성 : 한 속성에는 한 가지의 값만

> 특징을 모두 지니면 일종의 무결성

`-` 키의 종류

* 슈퍼키 : 튜플의 유일성을 구분짓는 키(단일 속성 내지 다차원 속성이 가능)
* 후보키 : 유일성 및 (차원)최소성 만족
* 기본키 : 후보키 중 기본으로 사용하는 하나의 키
* 대체키 : 기본키로 선택되지 못한 후보키

**기본키 ∈ 후보키 ⊂ 슈퍼키**

> 널 값을 가지면 안됨 / 자주 변경되면 안됨 / 단순해야함

* 외래키
> 릴레이션들 간의 관계에서 파생되는 키, 관계를 나타내는 릴레이션에 존재하며 각 릴레이션의 기본키를 가져온다.
>
> 기본키와 도메인이 같음.
> 
> 널값을 가질수도 있음.



> 순환관계에서의 외래키

### 관계 데이터 모델의 제약

`-` 개체 무결성 제약

* 기본키를 구성하는 모든 속성은 널값을 가지면 안됨 : 널값을 가진 튜플들이 구분되지 않음

`-` 참조 무결성 제약

* 외래키는 반드시 참고하는 기본키가 가지는 값을 가져야 함(기본키와 도메인이 같아야 함.)
* 널값을 가질 수 있음.
* 기본키를 수정하려면 외래키를 고려하거나, 외래키의 값을 삭제 또는 널값으로 지정하여 무결성 제약 만족 필요.

> 고객이 사이트를 탈퇴했을 때 : 기본키 ID가 탈퇴로 인해 삭제됐을 경우, 같은 외래키를 가지는 릴레이션 인스턴스를 삭제하거나, 외래키를 널값으로 지정해야 한다.

### 무결성

중복이 없는 것

기본키에는 널값이 없는 것

외래키는 참조하는 기본키가 가지는 값 또는 널을 가져야 하는 것
