### DBMS(데이터베이스 관리 시스템) 역사
Data Base Management System

* 제 1세대 DB 관리 시스템

> 그래프 구조로 데이터를 저장 : IDS
>
> 트리 구조로 데이터를 저장 : IMS

* **제 2세대 DBMS**

테이블로 구성하는 방식, SQL이 관계 DBMS에 특화되어 있음.

> 열 구성에 있어서 관련성이 떨어지는 정보는 최대한 피함(관련성이 있는 항목끼리 테이블을 구성함)
>
> 테이블 간 상호작용이 필요할 경우 키를 통해서 연결(포인터?)

* 제 3세대 DBMS

객체지향-객체관계 DBMS

> 객체 지향 : 인덱스만 지정해두고 그대로 데이터를 저장하는 방식(비정형데이터를 그대로 저장함)
>
> 객체 관계 : 객체지향과 관계 DBMS의 개념을 통합

* 제 4세대 이후 DBMS

관계 DBMS가 아닌 객체 지향으로 하되, 안정성을 가지도록. 오픈소스가 많아 보안 문제를 야기할 수 있음.

> NoSQL, NewSQL : 비정형데이터 처리를 전문적으로


**핵심적으로 알아볼 것은 2세대 DBMS**


### 정보와 데이터베이스 구조

* DB는 데이터를 쌓아놓은 것이며, 그 관리시스템은 분석 가공을 효율적으로 할 수 있는 기본 틀을 제공함.

내용기반 참조를 잘 할 수 있어야 함(내용을 기반으로 데이터를 불러오기 편해야 함.)


가장 먼저 해야할 것 : 데이터베이스 설계

* 이름 및 스키마 지정
> 저장할 열과 열마다 저장될 데이터 구조와 제약조건을 지정. (INT, FLOAT, CHAR(#)) / Integer, Float의 경우 비트 수 지정

* 3단계 데이터베이스 구조 : 외부 단계 / 개념 단계 / 내부 단계

> 개념 단계 : 전체 관리자의 관점에서 관리하는 단계, 기본적인 데이터 테이블의 구조(데이터 테이블을 만드는 것)
> 외부 단계 : 일반 이용자가 사용할 수 있는 단계, 특정 부분만 집중하여 관리
> 내부 단계 : 데이터베이스 시스템 전체의 하드 / 소프트웨어


`-` 개념 단계 : 모든 게 하나로 통합되어 있음, 하나의 스키마가 존재함.

`-` 외부 단계 : 사용자에게 편의성을 제공하는 것이 중점, 부분이 전체이므로 각각 겹치는 부분이 존재해야 한다. 외부 스키마는 다를 수 있다.(각 사용자가 보는 데이터 구조가 다를 수 있다.) 외부 스키마는 변동이 가능.

`-` 내부 단계 : 실제 저장장치의 관점


개념 단계에서 사용하는 이름 / 외부 단계에서 사용하는 이름이 달라도 됨 (스키마가 서로 달라도 됨), 단, 상호 공유되는 데이터에 대해서는 매핑이 되어야 함.


* 데이터 독립성 (Data Independency)

개념을 고정시키고 외부 스키마를 유동적으로 변화시켜야 함 : 매핑을 통해서 각 단계의 독립성을 유지하고자 한다.

> 내부 개념에서는 스키마의 이름이 중복되지 않는 것이 중요하나, 외부 단계의 데이터에서는 사용자의 편의성을 위해 유동적으로 중복을 허용하며 조작될 수 있음. 이로써 단계 간 독립성을 유지할 수 있음. 매핑, 일종의 사상으로 내부 단계와 외부 단게를 연결.

개념/내부 사상 | 외부/개념 사상

사상에 대한 정보를 어딘가 저장을 해야 함. Data dictionary / Meta data


1. 단계별 스키마 지정
2. 정의된 스키마들의 연결 : Meta Data, 외부에 공개되면 안되는 데이터


* 데이터베이스 사용자

> 데이터베이스 관리자 : 데이터베이스 시스템 DBMS을 운영하고 관리(개념 단계 / 내부 단계 총괄 관리 | 보안 및 운영)
> 최종 사용자(일반 사용자) : 데이터 이용 또는 분석
> 응용 프로그래머 : DB와 연동하여 분석하는 응용 프로그램을 작성, 알고리즘 등 제작


* 데이터 언어(내부 단계 개념)

> 데이터 정의어(DDL) : 스키마 정의ㆍ수정ㆍ삭제
> 데이터 조작어(DML) : 데이터 조작
> 데이터 제어어(DCL) : 응용프로그램 작성


`-` 데이터 정의어(DDL) : 새로운 데이터베이스 구축을 위한 스키마 정의 / 기존 스키마 삭제 및 수정 -> 데이터 사전에 저장됨

`-` 데이터 조작어(DML)

`-` 데이터 제어어(DCL) : DB내 데이터 무결성과 일관성 유지를 위해 내부적 필요규칙이나 정의를 하는데 사용한 언어.
> 무결성 : 정확하고 유효한 데이터 유지. 인덱스에 이빨이 빠져있는 경우, 결측치가 null이 아닌 등 제약조건을 위배
> 보안 / 회복 / 동시성의 성질을 유지토록 함.

* 데이터 사전 / 데이터베이스
> 데이터 사전의 경우 단계별 스키마들의 사상을 나타내는 자료로 보안이 중요




### 데이터 모델링

개념 단계에서는 데이터 테이블을 만드는 것이 목적.

> 테이블은 몇개가 필요한가, 테이블의 구조는 어떠한가, 스키마 구성 및 테이블 간 구조 등을 정리 : 데이터 모델링


ex) 수업에 참여하는 학생들에 대한 정보 : 강의실에 들어온 시간 / 강의실을 나간 시간 / 접속한 웹사이트(반응변수?)
-> 기타 관련된 여러가지 정보들을 취합 / 연결

* 데이터 구조 / 연산 / 제약조건

> 연산 : 테이블로 가는 관계가 어떻게 구성되는지 관계에 대한 연산 (메타데이터?)

`-` 개념적 데티어 모델 : 개체-관계 모델
> 개체 : 개념적으로 정의된 하나의 객체, 개별 데이터, 구별되는 모든 것
> 관계 : 개체 간 연결

**중요**
개체(entity) : 행, 구별되는 모든 것(단독이나 묶음, 사람이나 사물), 물리적인 존재만이 아닌 개념적 존재(과목, 학과) 가능
> 개체의 인스턴스(instance) : 행의 한 데이터 값.
속성(attribute) : 열, feature, 개체의 고유한 특성. 의미가 있는 가장 작은 데이터의 논리. 항목 간 호환이 되지 않는 사항이 없도록 가능한 작게 잡아야 함(어디까지 작아야 하는지는 정답이 없음.)
**중요**

* E-R 다이어 그램
개체와 속성을 그림.

> 속성의 분류
> - 단일 값 속성 / 다중 값 속성 : 여러 정보를 포함하느냐, 않느냐 (튜플이나 벡터로 표현된 정보 등)
> - 단순 속성 / 복합 속성 : 쪼갤 수 없느냐, 있느냐
> - 유도 속성 : 기존 속성의 값에서 다른 속성의 값이 유도되는 것, 통합하여 연동시킬 수 있음


`-` 논리적 데이터 모델 : 관계 데이터 모델
