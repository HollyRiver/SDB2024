#### 데이터 프레임을 csv 파일로 저정함

```python
d1 = pd.DataFrame({'key':['f','g','h'], 'data1':[3,4,6]})
d2 = pd.DataFrame({'key':['a','g','d','g','h','h'], 'data2':[1,4,5,7,9,10]})
print(d1)
print(d2)
pd.merge(d1,d2)
```

#### 데이터 프레임 두 개를 병합시킴 / 여기에서 같은 변수 이름을 가진 key 기준으로 이루어짐을 알 수 있음

```python
d1 = pd.DataFrame({'key':['f','g','h'], 'data1':[3,4,6]})
d2 = pd.DataFrame({'key':['a','g','d','g','h','h'], 'data2':[1,4,5,7,9,10]})
print(pd.merge(d1,d2))

print(pd.merge(d1,d2, left_on='key', right_on='key', how='right'))
print(pd.merge(d1,d2, left_on='key', right_on='key', how='left'))
print(pd.merge(d1,d2, left_on='key', right_on='key', how='outer'))

d2 = pd.DataFrame({'key2':['a','g','d','g','h','h'], 'data2':[1,4,5,7,9,10]})
print(pd.merge(d1,d2, left_on='key', right_on='key2', how='outer'))
print(pd.merge(d1,d2, left_on='key', right_on='key2', how='inner'))
```

#### 병합 명령문을 이용하여 조인들을 구현하는 구문 <br>
순서대로 내부 조인, 오른쪽 외부 조인, 왼쪽 내부 조인, 전체 외부 조인 그리고 마지막 구문은 전체 외부 조인을 하되 두개의 변수를 비교해서 조인을 하고 두 개의 변수가 하나로 병합되지 않음 
#### 특별히 마지막 조인을 세타 조인이라고 부름: 특정변수들의 동일성이나 대소 관계를 조건으로 할 수 있음  

```python
s1 = pd.Series([1,2], index=['A', 'B'])
s2 = pd.Series([2,4,5], index=['A', 'D', 'E'])
s3 = pd.Series([6,7], index=['F', 'G'])

print(pd.concat([s1,s2],axis=0))
print(pd.concat([s1,s2],axis=1))
pd.concat([s1, s2, s3], keys=['one', 'two', 'three'])
```

#### 자료를 열이나 행으로 붙이고 있음 / 또한 행으로 붙였을 때, 어떤 객체에서 왔는가를 명기

```python
pd.concat([s1, s2], join='inner', axis=1, keys=['one', 'two'])
```

#### 중복 여부 확인
