### 중간고사 리뷰

`-`

1) 데이터베이스 정의와 특성(성질)
2) 스키마 : 변수명과 속성에 포함되는 값의 범위
3) 데이터 사전과 내부, 개념, 외부단계
4) 기본키가 만족해야 할 조건
5) 데이터 모델의 구성
6) 튜플의 유일성

`-`

1) 비정형데이터도 수치화가 된 것임
2) 관계 DBMS는 2세대
3) 외래키는 널을 사용할 수 있음(참조 무결성)
4) 릴레이션에서 튜플들의 순서는 의미가 없음(튜플의 무순서)
5) 관계대수에서 카티션 프로덕트는 튜플들의 속성이나 차수가 달라도 연산이 가능
6) 조인을 하는 릴레이션의 카디널리티가 각각 3과 7일 때, 자연조인하면 카디널리티는 10이하이다. -> 전부 중복된다고 하면 21까지 가능하다.

`-` E-R 다이어그램

1) 네모는 개체, 마름모는 관계
2) 개체는 무조건 릴레이션을 만들고, 관계는 다대다라면 무조건 만들고 일대다라면 만들수도 있다. -> 관계의 릴레이션을 만들어두고, 한쪽에 붙여놓는 게 효율적일 때가 있음.
3) 이중마름모는 종속관계

`-`

1) 릴레이션이 되려면 중복되는 튜플이 없고, 속성의 값이 분할될 수 없어야 함. 또한 중복되는 테이블이 없어야 함.
2) 기본키는 튜플들을 관리하기 위해 사용 : 원하는 튜플들을 불러오거나 삭제.

> 기본키 = 3로 쿼리문을 입력하면, 하나의 튜플만 나와야 한다. (도메인을 준수한다는 가정 하에.)

3) 후보키가 하나밖에 없으면 대체키는 존재하지 않는다.(원칙적으로)

`-` 카티션프로덕트도 릴레이션을 생성하므로 튜플의 유일성을 만족한다.
