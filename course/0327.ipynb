### 관계 데이터 연산

데이터 모델의 구성

* 데이터 구조 : E-R diagram, 스키마
* 제약조건 : 일관성, 무결성
* 연산 : 처리 요구(query)

`-` 일반집합 연산자

* 합
* 교
* 차 -> 위 셋은 차수가 같아야 함, 포함하고 있는 정보가 달라도 도메인만 같으면 일반 집합 연산이 가능
* 곱(카티션 프로덕트) -> 차수와 스키마 모두 달라도 됨

`-` 순수 관계 연산자

* select : 튜플을 고름. $\sigma_{Restrictions}(A)$, Restrictions를 predicate라고 함(비교식)
* project : 속성을 고름. $\pi_{Attributes}(A)$
* join : 속성의 같은 값을 기준으로 병합 -> 동등 조인(키를 둘다 살림) / 자연 조인(키 중 하나만 살림)(`merge(inner)`) $A ⨝_{a=b} B~/~A ⨝_N B$
* division : 속성의 같은 값을 기준으로 한쪽 테이블만 반환 $A ÷ B$

> 조인 : 탈퇴한 고객의 경우 구매 내역 테이블에는 외래키가 널인 상태로 들어가있음. $\to$ 병합할 때 자연조인을 하면 구매 내역 테이블이 고객 정보는 널인 상태로 튜플로 붙는다.

`-` 정보 유지와 조인

> 동일한 내용이 2개 / 3개의 튜플이 있으면 정보를 유실시키지 않도록 카티션 프로덕트처럼 곱해서 6개의 튜플로 저장한다.

`-` semi-join

> 자연조인을 하되, 한 쪽의 릴레이션에서만 정보를 가져옴.



`-` outer join (`merge(left, right, outer)`)

> 자연조인을 하되, 릴레이션의 정보를 전부 남겨둠. 존재하지 않는 속성은 NULL로 처리

**정보가 모두 겹친다면, 자연조인과 외부조인이 같음**

`-` 기본키가 아닌 것으로 조인을 하는 경우

> 외래키로 조인(NULL 존재) : 조인하는 속성의 값도 널로 넣을 수 있음.


> 전체 외부 조인

* $\theta$ join : 특정한 조건을 만족하게 하는 조인. 카티션프로덕트 후 셀렉트하는 조인.

> $R \Join_{R.ID < S.ID} S \to$ R의 ID의 값 하나에 대하여 그보다 큰 모든 S의 튜플을 붙여서 만듦
