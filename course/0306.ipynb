### 데이터베이스의 정의 : 여러 사용자가 공유하여 사용할 수 있도록 통합하여 저장한 운영데이터의 집합

* 공유데이터 : 여러 사용자가 동시에 사용할 수 있음
* 통합데이터 : 데이터의 중복은 허용하지 않거나 최소화함
* 저장데이터 : 컴퓨터로 접근 가능한 저장방식 사용
* 운영데이터 : 조작 및 이용을 위한 지속성 존재


### 데이터베이스의 기본 특징

* 실시간 접근성 : 언제나 접근이 가능함
* 계속 변화 : 내용이 계속 수정 및 업데이트됨
* 내용 기반 참조 : 내용을 기반으로 데이터 탐색이 가능함
* 동시 공유 : 여러 사용자가 동시에 이용 가능함


### 데이터의 분류

`-` 형식
* 정형데이터(태뷸러데이터)
* 반정형데이터
* 비정형데이터

`-` 특성
* 범주형 / 수치형
* 이산형 / 연속형
* 명목 / 순서 / 이산 / 연속


### 데이터베이스 관리 시스템

**파일에 대한 효율적인 관리 및 접근을 위해 관리시스템이 요구됨**

정수형으로 데이터 저장 시 문자보다 리소스 사용양이 적음

데이터베이스 : 하드웨어와 소프트웨어가 동시에 필요

DB 제작을 위한 오픈소스가 많아지고 있음



* 데이터베이스 관리 시스템

기존 자료 저장 방식 : 파일 시스템
> 내용에서 중복이 있으며, 이는 저장공간의 낭비, 일관성 유지의 어려움(잘 처리가 되면 좋으나, 중복된 정보를 실수로 모두 바꾸지 못할 경우 해당 시스템 사용이 어려워짐), 무결성 유지의 어려움(데이터 저장의 규칙 자체를 비틀 수 있음)을 가지게 됨.

위 세가지 문제를 해소하기 위해 DB를 이용한다.

* 파일 시스템 극복

> 중복된 정보를 따로 저장해두고 참조의 형식으로 달아둠(ref)
> 
> 

* 데이터 종속성(중복된 정보를 포함하는 개별 데이터 간 종속성)

> 프로그램 접근 중 데이터 항목을 늘리는 등에서 관련 파일 두 개에 대해 동시 처리해야 하므로 복잡성이 증가
> > 데이터 파일의 동시 공유, 보안, 회복 기능의 약화
> > 프로그램 개발을 어렵게 만듦

최근 DB > 그냥 여러곳에 중복되는 방식으로 저장함. --> 복구할 때 전부 복구해야 하므로 어려움이 증가

* 데이터베이스 관리 시스템

정의 기능 : 필요한 데이터를 데이터베이스에 통합적으로 저장하고 관리를 집중적으로 담당 > 스키마
조작 기능 : 데이터베이스에 들어있는 데이터 삽입, 삭제, 수정, 탐색 및 응용프로그램이 데이터베이스를 공유하게 함 > 종속성 때문에 어려움이 있음
제어 기능 : 데이터를 정확하고 안전하게 유지, 보안과 사고 후 처리도 포함

> 정의와 조작 기능을 중점적으로 강의할 것

* 데이터베이스 관리 시스템

장점
> 데이터 중복성 회피 : 저장공간 절약 및 일관성 유지
> 데이터 독립성 : 중복되지 않아서 응용프로그램이 영향을 받지 않음
> 데이터 동시 공유 : 여러 응용프로그램 지원, 통신 이슈 등으로 데이터 간 싱크가 맞지 않을 경우 내용이 달라지면서 문제 발생 가능
> 데이터 보안 향상
> 데이터 무결성 유지
> 표준화가 가능
> 장애 발생 시 회복 가능
> 응용프로그램 개발 비용 감소

단점
> 비용의 증가 : 중앙집중식과 저렴한 다수의 서버
> 백업, 회복이 복잡 : 여러 사용자 동시 지원으로 장애의 원인과 이유 파악이 어려움, 백업도 복잡
> 중앙집중 관리로 인한 취약점 : 시스템 장애가 전체 업무 처리의 중단으로 이어짐

따라서 데이터가 자주 바뀌지 않거나, 소규모인 경우 데이터베이스 관리 시스템이 없는 것이 나을수도 있음.
