`-` **스키마** : 데이터베이스에 저장된 데이터 구조와 제약조건

> 제약조건을 너무 많이 붙여놓으면 사용이 복잡해짐, 제약조건이 너무 없으면 구조상 문제 발생할 수 있어...

`-` 3단계 데이터베이스

* 외부단계 : 사용자 관점
* 개념단계 : 
* 내부단계 : 저장장치, 하드/소프트웨어 관점
> 단계별 독립성(개별 변경 시 다른 단계가 바뀌면 안됨) 유지가 중요

`-` 데이터베이스 사용자

* 관리자 : 개념 / 내부단계 설계 및 유지, 보안
* 최종사용자 : 데이터베이스의 자료를 사용
* 응용프로그래머 : 데이터베이스에 연동되는 응용프로그램을 설계

> 데이터 사전과 데이터베이스. 데이터 사전 : 데이터 딕셔너리, 카탈로그, 메타 데이터. 스키마에서 스키마로의 사상에 대한 정보를 가지고 있음.

### 데이터 모델링

1) 데이터 구조
2) 연산
3) 제약조건

을 설계하는 것. (데이터 구조 + 제약조건 = 스키마)

`-` 대표적인 방법

* 개념적 데이터 모델 : 개체-관계 모델
* 논리적 데이터 모델 : 관계 데이터 모델


### 개체-관계 모델

개체(entity)와 개체 간의 관계를 이용해 현실세계를 개념적 구조로 표현 $\to$ 개체-관계 다이어그램



> 관계에서도 데이터가 축적됨.

* 개인(entity)과 학번 : 개인과 학번이 일대일 관계를 가질 수 있다. 따라서 관계 데이터를 따로 두는 게 아닌 학번을 그대로 이어버리는 게 좋을 수 있다. (단순한 관계)
* 고객과 매장 : 고객이 여러 매장과 관계를 가질 수 있다. 관계를 나타내는 테이블을 따로 만드는 게 편할 수 있다. (복잡한 관계)

> 따라서 관계에 대한 분류가 필요하다.


* 할 수 있다면 테이블은 적게 있어야, 따라서 많은 인스턴스(행)를 가지도록 하는 게 바람직함.

개체 : 고객(이름, 주소, 연락처, 적립금) / 개체 인스턴스 : <정소화, 부천시 원미구, 010-666-666, 2000>

속성 : 고객아이디, 고객명


* 속성 값의 개수에 따라 : 단일 값 속성 / 다중 값 속성
* 의미의 분해 가능성에 따라 : 단순 속성 / 복합 속성
* 기존 속성 값에서 유도되는 경우 : 유도 속성 > 학생이라는 개체가 있으면, 학생증이 유도됨

`-` 단일 값 속성 / 다중 값 속성

> single-valued attribute : 금액(원 단위). 하나의 값만 가짐
>
> multi-valued attribute : 연락처. 집전화번호, 휴대전화번호 등 다중값이 가능 (이중 원으로 표시)

`-` 단순 속성 / 복합 속성

> simple attribute : 더는 분해가 안되는 속성
>
> conposite attribute : 의미가 분해되어 여러 개의 값을 가짐. 주로 쪼개어 관리됨

`-` 유도 속성

> derived attribute : 기존의 다른 속성값으로 결정되는 속성

`-` 널 속성

> null attribute : 결정이 안되거나 모르는 값. 허용을 안하면 좋겠는데, 허용할 수밖에 없는 상황이 있음.

`-` 키 속성

> key attribute : 개체 집합에 존재하는 각 개체 인스턴스를 식별하는 데 사용. 인덱스 속성. (글자에 줄을 그어 표시)
>
> - 이름, 나이, 성별, 거주 지역 등은 중첩이 되어 키 속성에 적합하지 않음.
>
> 인스턴스의 값이 동일하다면 인스턴스 간 구분이 되지 않음. 따라서 키 속성으로 인스턴스 간 독립성을 유지하는 것이 중요.

* 관계

`-` 개체와 개체 간의 연관성

> 개체와 개체 사이에서 정의되어 속성을 가질 수 있음. 마찬가지로 인스턴스를 가질 수 있음(관계 인스턴스).



> 아래에 속성을 파생시킬 수 있다.

`-` 일대일 : 개체의 인스턴스 하나 당 관계는 하나만 가질 수 있다.(관계를 맺지 않는 개체가 있을 수 있음, 다 맺는다면 수도 속성)

`-` 일대다 : 한 쪽의 개체에서는 여럿의 관계를 맺을 수 있음.

`-` 다대다 : 서로 여럿의 관계를 맺을 수 있음.

> 맺을 수 있는 관계가 무한하진 않음.
