**Mongo DB, 비정형 DB의 기본**

1. Key-Value : 키와 값으로 구성된 가장 단순한 구조
2. Document : Json / Xml

* Non-SQL 데이터베이스
    * Tensor : 다차원 배열, 이미지와 같은 데이터를 수치로 표현.
 > 내용이나 데이터의 구조를 유지하기 위해 존재. 3차원 텐서 : 이미지, 4차원 텐서 : 영상.
> 영상 $\to$ speach, sound, captions 등 여러 정보

    * 유연성 : 고정되지 않은 스키마로 인하여 비정형 데이터 저장에 유리, 반복적이고 빠른 실행 가능
    * 확장성 : 분산저장으로 인해 수평적 확장이 가능
    * 고기동, 고성능, 빠른 검색 및 데이터 추가

* Non-SQL 데이터베이스 유형
    * Column : 2차원 행렬로 펴서 저장, null값이 생길 수 있음. **좋은 구조가 아님**
    * Key-Value : Json, Xml과 같은 구조로 저장.
    * Graph : 관계 노드들을 직접 연결시킴. 관리가 어려움.  **선호되지 않는 구조**
    * Document : 트리형 자료구조로 데이터를 저장함. 위계를 만들어 자료 탐색에 유리.
      > Key-Value / Document를 기본적으로 활용한다.

### Document

`-` Key와 Value로 구성되어 있으며, 값은 Document로 대부분 Json, Xml, YXml과 같은 인코딩 방식을 사용.

> Json의 경우 Python에서의 dictionary 구조와 동일.

`-` Wide-column : 하나의 행에 길이가 다른 여러 개의 열을 가질 수 있음, Graph : 각각의 개체와 관계를 저장

ex) Mongo DB



> Relation = Collection / tuples = Documents / Attributes = Fields

replica set

P-S-S / P-S-A 방식으로 보조 데이터셋을 운용, 이상 발생 시 메인으로 변환

Parent / Array of ancestors(부모 전체) / child references / Material path(자식 전체)

Nested sets : 검색 경로를 따로 지정

### 관계 데이터 연산

데이터 구조 / 제약조건 / 연산

* 데이터 구조 : E-R diagram, 스키마 등
* 제약조건 : 일관성(접근하는 사용자들에게 동일한 데이터가 전달), 무결성(개체 무결성 제약 : 기본키는 널값을 가질 수 없음 / 참조 무결성 제약 : 외래키가 가질 수 있는 값의 범위는 기본키를 벗어날 수 없고, 중복을 허용하며, 널값을 가질 수 있음)

* **연산** : 원하는 데이터를 얻기 위해 릴레이션에 필요한 처리 요구를 수행
  > 데이터에 대한 처리 요구 : 질의(query)

`-` 관계 대수와 관계 해석

* 관계 대수의 개념 및 연산자
    * 릴레이션을 처리하는 과정을 순서대로 기술한 연산자들의 집합
     > 연산자들의 대상 : 릴레이션, 결과 : 릴레이션
     
    * 일반집합 연산자
      > 합집합 ∪ / 교집합 ∩ / 차집합 － / 카티션 프로덕트(cartesian product) ×(릴레이션 간 모든 순서쌍을 만들어내는 것)
      >
      > 연산을 해도 릴레이션이 나온다. (세 가지 유일성 / 무순서 / 원자성 만족하는 게 나와야 함) 그래야만 관계적으로 완전한 DBMS임
      
    * 순수 관계 연산자
      

select는 튜플을 슬라이싱, project는 몇 개의 속성만 슬라이싱하는 것(정보 수를 줄이는 것이니까 사영)

join은 도메인이 같은 속성을 이용하여 merge하는 것. 도메인만 같으면 됨.

> 속성의 값이 같은 행들만 연결하여 반환하는 것을 가장 먼저 생각해볼 수 있음. 속성의 이름이나 스키마가 달라도 도메인만 같으면 가능함. $\to$ 속성 이름이 다를 경우 두개를 다 쓰거나 하나만 쓸 수 있고, 새로 지정도 가능
>
> 한 쪽의 값만 있는 행이 있으면, 반환하는 릴레이션에서는 결측치를 null로 집어넣는 것을 고려해볼만 함.

* 각 속성을 가져올 때, ID1과 ID2가 있다면 둘 중 하나를 없애고(중복되니까) 남은 속성의 이름을 ID로 짓는 것 : natural join / 자연 조인

division은 도메인이 같은 속성에 대하여 릴레이션 간 정보가 겹치는 내용에 해당하는 이전 릴레이션의 튜플들을 반환

연산을 위해서는 2개의 피연산자가 필요(당연)

* 일반 집합 연산
> 합/교/차 : 피연산자인 2개의 릴레이션이 합병 가능해야 함
>
> 1. 두 릴레이션의 차수가 같아야 함
> 2. 서로 대웅되는 속성의 도메인이 같아야 함
>
> 카디션 프로덕트 : 두 릴레이션의 속성 간 값이 같은 행만 남겨두면, 조인과 같은 결과를 낳는다.
>
> * 다른 거여도 상관이 없다... 그냥 늘여서 튜플을 나열하면 됨

* 순수 관계 연산자

> select : ∧ ∨ ¬ and or not
>
> project : 해당 속성만 선택
>
> join : 동등 조인(두 속성을 남겨둠) / 자연 조인
>
> * 키 속성이 아닌 다른 속성으로 조인을 할 경우 중복되는 값이 나올 수 있음. 이 경우 값이 같은 것들을 모두 엮어 카티션 곱과 같이 연산해야...
>
> division : R ÷ S, R에 있는 속성들 중 S와 관련 있는 속성들로 릴레이션을 반환. S에 대한 정보는 제거하여 반환함.

semi-join : 한 쪽의 정보(뚫린 방향)는 가져오지 않는 것
